# Bar plot

### Creating a Bar plot for the comparison of RMSE values for different models 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams.update({'figure.facecolor':'white'})
```

```python
#df = pd.read_csv('data/Models_Aany.csv')
```

```python

labels = ['Baseline', 'SARIMAX', 'Prophet', 'NeuralProphet']
NH = [1.361, 1.424, 0.519, 0.465]
Sea_of_Okhotsk = [1.216, 0.083, 0.087, 0.086]
Hudson_Bay = [1.204, 0.171, 0.089, 0.090]
Central_Arctic = [1.002, 0.178, 0.100, 0.096]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(15, 8))
rects1 = ax.bar(x - width/2, NH, width, label='NH', color='slategrey')
rects2 = ax.bar(x + width/2, Sea_of_Okhotsk, width, label='Sea of Okhotsk', color='cornflowerblue')
rects3 = ax.bar(x + width/2 + width, Hudson_Bay, width, label='Hudson Bay', color='slateblue')
rects4 = ax.bar(x + width/2 + width + width, Central_Arctic, width, label='Central Arctic', color='darkblue')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('RMSE')
#ax.set_title('Model Performance')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()

plt.show()
fig.savefig('images/RMSE_Plot.png', dpi=300)
```

```python
labels = ['NH', 'Sea_of_Okhotsk', 'Hudson_Bay', 'Central_Arctic']
Baseline = [1.361, 1.216, 1.204, 1.002]
SARIMAX = [1.424, 0.083, 0.171, 0.178]
Prophet = [0.519, 0.087, 0.089, 0.100]
NeuralProphet = [0.465, 0.086, 0.090, 0.096]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(15, 8))
rects1 = ax.bar(x - width/2, Baseline, width, label='Baseline', color='slategrey')
rects2 = ax.bar(x + width/2, SARIMAX, width, label='SARIMAX', color='cornflowerblue')
rects3 = ax.bar(x + width/2 + width, Prophet, width, label='Prophet', color='slateblue')
rects4 = ax.bar(x + width/2 + width + width, NeuralProphet, width, label='NeuralProphet', color='darkblue')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('RMSE', fontsize=15)

#ax.set_title('Model Performance')
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=15)

ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    fontsize=12,
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()

plt.show()
fig.savefig('images/RMSE_Plot2.png', dpi=300)
```
